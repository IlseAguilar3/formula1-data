```python
#Hasta 23 porque el último número es cota y no lo toma en cuenta y queremos las 22 carreras del 2022
list(range(1,23))
```

```python
#Ciclo que extrae todas las carreras del 2022, irá extrayendo uno x uno, pero al final sólo podremos ver el último

df_results_2022 = pd.DataFrame() #el dataframe contendrá los json de cada carrera
for race in list(range(1,23)):
    url = 'https://ergast.com/api/f1/2022/{}/results.json'
    response = requests.get(url.format(race)) #.format() indica que la variable race es la que irá cambiando y debe asignarle el formato
    content = json.loads(response.content)
    df_results_2022 = df_results_2022.append(pd.DataFrame(content['MRData']['RaceTable']['Races'][0]['Results']))
```

```python
df_results_2022
```

```python
#convertimos la columno de points a enteros
df_results_2022['points'] = df_results_2022.points.astype(int)
```

```python
df_results_2022['Driver'] = df_results_2022['Driver'].apply(lambda x: x['driverId'])
```

```python
df_results_2022.head()
```

```python
#total de puntos realizados por piloto en las carreras del 2022
#método aggregate, hace que a los puntos les aplique una suma
df_results_2022.groupby(['Driver']).agg({'points':'sum'}).sort_values(by='points', ascending = False)
```

```python
df_results_2022['Constructor'] = df_results_2022['Constructor'].apply(lambda x: x['constructorId'])
```

```python
df_results_2022.head()
```

```python
#total de puntos realizados por equipo en las carreras del 2022
#método aggregate, hace que a los puntos les aplique una suma
#31/01/23
df_results_2022.groupby(['Constructor']).agg({'points':'sum'}).sort_values(by='points', ascending = False)
```

## Tabla de races

```python
#Creación del diccionario vacío para incorporar todos los datos de las carreras del 2010:2022
races={'season':[],
      'round':[],
       'raceName':[],
       'circuitId':[],
       'lat':[],
       'long':[],
       'country':[],
       'date':[],
       'url':[]}
```

```python
#Ciclo para enrobustecer nuestro diccionario de races
for year in list(range(2010,2023)):
    url = 'https://ergast.com/api/f1/{}.json'
    response = requests.get(url.format(year)) #.format() indica que la variable race es la que irá cambiando y debe asignarle el formato
    content = json.loads(response.content)
    
    for item in content['MRData']['RaceTable']['Races']:
        races['season'].append(int(item['season']))
        races['round'].append(int(item['round']))
        races['raceName'].append(item['raceName'])
        races['circuitId'].append(item['Circuit']['circuitId'])
        races['lat'].append(float(item['Circuit']['Location']['lat']))
        races['long'].append(float(item['Circuit']['Location']['long']))
        races['country'].append(item['Circuit']['Location']['country'])
        races['date'].append(item['date'])
        races['url'].append(item['url'])
```
